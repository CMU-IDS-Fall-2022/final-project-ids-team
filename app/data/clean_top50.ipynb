```python
df = pd.read_csv("top10s.csv", encoding='latin-1')




changer = {}
changer['detroit hip hop'] = 'hip hop'
# changer['canadian pop'] = 'pop'
changer['candy pop'] = 'other pop'
changer['barbadian pop'] = 'other pop'
changer['baroque pop'] = 'other pop'
changer['colombian pop'] = 'other pop'
changer['danish pop'] = 'other pop'
changer['folk-pop'] = 'other pop'
changer['acoustic pop'] = 'other pop'
changer['french indie pop'] = 'other pop'
changer['metropopolis'] = 'other pop'
changer['moroccan pop'] = 'other pop'
changer['art pop'] = 'other pop'
changer['indie pop'] = 'other pop'
changer['australian pop'] = 'other pop'
changer['atl hip hop'] = 'hip hop'
changer['hip pop'] = 'hip hop'
changer['canadian hip hop'] = 'hip hop'
changer['australian hip hop'] = 'hip hop'
changer['canadian contemporary r&b'] = 'r&b'
changer['alternative r&b'] = 'r&b'


changer['electro house'] = 'house'
changer['tropical house'] = 'house'

changer['edm'] = 'electro'
changer['electronic trap'] = 'electro'

changer['permanent wave'] = 'other'
changer['latin'] = 'other'
changer['belgian edm'] = 'other'
changer['downtempo'] = 'other'
changer['brostep'] = 'other'
changer['escape room'] = 'other'
changer['chicago rap'] = 'other'
changer['celtic rock'] = 'other'
changer['alaska indie'] = 'other'
changer['hollywood'] = 'other'
changer['irish singer-songwriter'] = 'other'
changer['canadian latin'] = 'other'
changer['contemporary country'] = 'other'

def change_genre(x):
    if x in changer:
        return changer[x]
    else: 
        return x
    
df["Top Genre"] = df["Top Genre"].apply(lambda x: change_genre(x))

a = df["Top Genre"].tolist()

dic = {}
for it in a:
    if it not in dic:
        dic[it] = 0
    dic[it] += 1
    

dic = {k: v for k, v in sorted(dic.items(), key=lambda item: item[1], reverse = True)}
print(dic)
for k in dic:
    print("changer['"+k+"'] = 'other'")
```
